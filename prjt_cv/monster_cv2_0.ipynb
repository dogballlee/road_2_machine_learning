<font color=black size=5 face=雅黑>**傅里叶变换**</font>

*傅里叶变换的作用*：  

高频：变化剧烈的灰度分量，例如边界  
低频：变化缓慢的灰度分量，例如一片大海

*滤波*：  

低通滤波器：只保留低频，会使图像模糊  
高通滤波器：只保留高频，会使图像细节增强

在opencv中，通过cv2.dft()和cv2.idft()实现，输入图像要先转换成np.float32格式
得到的结果中频率为0的部分会在左上角，通常要转换到中心位置，可通过shift变换来实现
cv2.dft()返回的结果是双通道的（实部与虚部），通常还需要转换成图像格式才能展示(0,255)

```python
img_float32 = np.float32(imgGray_r)    #先转成float32，opencv的要求
dft = cv2.dft(img_float32,flags = cv2.DFT_COMPLEX_OUTPUT)    #返回的结果是双通道的（实部，虚部），通常还需要转换成图像格式才能展示（0,255）
dft_shift = np.fft.fftshift(dft)    #将得到的结果中频率为0的部分从左上角转换到中心位置

rows,cols = imgGray_r.shape
crow,ccol = int(rows/2),int(cols/2)    #中心位置

magnitude_spectrum = 20*np.log(cv2.magnitude(dft_shift[:,:,0],dft[:,:,1]))
#上面是一个映射公式，可以把之前做转换的图重新转换成灰度图能表示的形式 （频谱图）

#低频滤波掩模
mask_1 = np.zeros((rows,cols,2),np.uint8)
mask_1[crow-30:crow+30,ccol-30:ccol+30] = 1

#高频滤波掩模
mask_2 = np.ones((rows,cols,2),np.uint8)
mask_2[crow-30:crow+30,ccol-30:ccol+30] = 0


#IDFT逆傅里叶变换
fshift =dft_shift*mask_1
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:,:,0],img_back[:,:,1])    #低频滤波过程

fshifth =dft_shift*mask_2
f_ishifth = np.fft.ifftshift(fshifth)
img_backh = cv2.idft(f_ishifth)
img_backh = cv2.magnitude(img_backh[:,:,0],img_backh[:,:,1])    #高频滤波过程



plt.subplot(221),plt.imshow(imgGray_r,cmap='gray')
plt.title('input image'),plt.xticks([]),plt.yticks([])    #原图
plt.subplot(222),plt.imshow(magnitude_spectrum,cmap='gray')
plt.title('magnitude_spectrum'),plt.xticks([]),plt.yticks([])    #傅里叶变换后的频谱图
plt.subplot(223),plt.imshow(img_back,cmap='gray')
plt.title('img_back'),plt.xticks([]),plt.yticks([])    #再进行一次低频掩模逆变换后的图
plt.subplot(224),plt.imshow(img_backh,cmap='gray')
plt.title('img_back'),plt.xticks([]),plt.yticks([])    #再进行一次低频掩模逆变换后的图
plt.show()
```
